```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
weather_api_key = "1eaceb27a6d693fbca09e6aca3540841"

# Import API key
#from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units="imperial"
query_url = base_url + "APPID=" + weather_api_key + "&units=" + units + "&q="
#response = requests.get(query_url).json()
#response
query_url
```

```python
#from rubric 
#City latitude ,City longitude --> #lat +long in coord,
#Max temperature --> in main(temp_max)
#Humidity---> in main(humidity)
#Cloud coverage --> clouds--> all
#Wind speed--> wind
#City country--> sys and then country
#City datetime  --> dt(unix, UTC)
```

```python
citylist = []
lat = []
long = []
maxtemp = []
humid = []
cloud = []
wndspeed = []
country = []
date = []
```

```python
citycount = 0
counter = 1
print("Beginning Data Retrieval ")
print("-------------------------------")




# For each city name in cities list, do name,cloud,country,date,humid,lat,long,maxtemp,wndspeed
for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        citylist.append(response["name"])
        lat.append(response["coord"]["lat"])
        long.append(response["coord"]["lon"])
        maxtemp.append(response['main']['temp_max'])
        humid.append(response["main"]["humidity"])
        cloud.append(response["clouds"]["all"])
        wndspeed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        
        if citycount > 49:
            citycount = 0
            counter = counter + 1
    
        else:
            citycount = citycount + 1
            
        print(f"Processing Record {citycount} of Set {counter} : {city}") 
  
    except(KeyError, IndexError):
        print("City not found. Skipping...")

print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
```

```python
df1 = pd.DataFrame({ 
                "City" : citylist, "Latitude": lat, "Longitude": long, "Max Temperature": maxtemp,
    "Humidity": humid, "Cloud Coverage": cloud, "Wind Speed": wndspeed, "Country": country, "Date": date})
df1. head ()
```

```python
#creating a function for r- value/ correlation so I don't have to keep typing the same code out, removed plt.annotate
#as it was obscuring the graph too much, printed with r-value instead.
equ=[]
def correlation (x,y, xy=(0,30)):
    rvalue=st.pearsonr(x, y)[0]
    slope, intercept, rvalue, pvalue, stderr = st.linregress(x,y)
    rline = (x*slope) + intercept
    equation = 'y = ' + str(round(slope, 2)) + 'x + ' + str(round(intercept, 2))
    plt_size = plt.figure(figsize=(12, 8))
    plt.scatter(x, y)
    plt.plot(x,rline, color="red")
    plt.annotate(equation, xy=xy,textcoords="offset points", xytext=(0,10),ha='center')            
    rrvalue= round(rvalue,2)
    print(f"The r-value is :{rrvalue}")
    
   

```

```python
#Northern Hemisphere -Temperature (F) vs.Latitude
nhtl = correlation(north["Latitude"], north["Max Temperature"],(10,30))

plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("Northern Hemisphere - Max Temperature vs. Latitude Linear Regression")

             
plt.savefig("./output_images/Northern Hemisphere - Max Temperature vs. Latitude Linear Regression.png")
```

```python
#Southern Hemisphere - Max Temp vs. Latitude Linear Regression
shtl = correlation(south["Latitude"], south["Max Temperature"], (-40, 80))

plt.title("Southern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("Southern Hemisphere - Max Temperature vs. Latitude Linear Regression")
plt.savefig("./output_images/Southern Hemisphere - Max Temperature vs. Latitude Linear Regression.png")
```

### As latitude gets further from 0 the temperature decreases, this is because it is getting further away from the Equator line. The Sun's rays are less direct at higher and lower latitudes, which is why the North and South pole are so cold.

### Generally it seems that the humidity is highest at Latitude=0. Humidity will overall decrease as we move away from the zero point, however there are outliers to this, it's not a very strong correlation. Such thing as cold, humid environments can exist, large bodies of water can effect city humidity too. So it's not necessarily always the case that as you move away from the equator humidity will increase.
